```python
# Check for missing values
print(df[predictors].isnull().sum())
```

```python
# Impute missing values
df["OP100_Capuchon_insertion_mesure"] = df["OP100_Capuchon_insertion_mesure"].fillna(df["OP100_Capuchon_insertion_mesure"].mean())
```

```python
# Check for outliers
sns.boxplot(data=df[predictors])
plt.show()
```

```python
# Remove outliers
for predictor in predictors:
    q1 = df[predictor].quantile(0.25)
    q3 = df[predictor].quantile(0.75)
    iqr = q3 - q1
    df = df[~((df[predictor] < q1 - 1.5 * iqr) | (df[predictor] > q3 + 1.5 * iqr))]
```

```python
# Normalize the data
for predictor in predictors:
    if pd.api.types.is_numeric_dtype(df[predictor]):
        df[predictor] = df[predictor] / df[predictor].std()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression # fixed error with code in order to proceed with other steps
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, roc_auc_score
```

```python
# # Load the preprocessed data
# df = pd.read_csv("production_line_data_preprocessed.csv")
```

```python

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df[predictors], df[target], test_size=0.25)
```

```python
# Train five models
models = [LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier()]
```

```python
# Evaluate the models
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    print(model.__class__.__name__, f1, roc_auc)
```
