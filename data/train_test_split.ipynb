```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('valeo_dataset.csv')
df.head()
```

```python
X = df.drop(columns=['PROC_TRACEINFO','Binar_OP130_Resultat_Global_v'])
y = df['Binar_OP130_Resultat_Global_v']
```

```python
X_train, X_holdout, y_train, y_holdout = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
train_df = pd.concat([X_train, y_train], axis=1)
holdout_df = pd.concat([X_holdout, y_holdout], axis=1)
```

```python
train_df.to_csv('train.csv', index=False)
holdout_df.to_csv('holdout.csv', index=False)
```
